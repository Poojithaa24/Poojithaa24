```python
#without parameters and with return
def avg():
  a = int(input("Enter a value : "))
  b = int(input("Enter b value : "))
  c = int(input("Enter c value : "))
  return (a+b+c)/3

x = avg()
print("The average is : ",x)
```

```python
#with parameters and without return
def avg(a,b,c):     # a,b,c are formal parameters
  d =  (a+b+c)/3
  print("The average is : ",d)

x = int(input("Enter a value : "))
y = int(input("Enter b value : "))
z = int(input("Enter c value : "))
avg(x,y,z)          # x,y,z are actual parameters
```

```python
#with parameters and with return
def avg(a,b,c):
  return (a+b+c)/3

a = int(input("Enter a value : "))
b = int(input("Enter b value : "))
c = int(input("Enter c value : "))
x = avg(a,b,c)
print("The average is : ",x)
```

```python
#WPP for finding area of triange(sides given)
def ar(a,b,c):
  s = (a+b+c)/2
  area = (s*(s-a)*(s-b)*(s-c))**(1/2)
  return area

a = int(input("Enter first side of a triangle : "))
b = int(input("Enter second side of a triangle : "))
c = int(input("Enter third side of a triangle : "))
x = ar(a,b,c)
print("The area of the triangle is :",x)
```

```python
#WPP for finding roots of quadratic equations
def roots(a,b,c):
  d = b*b-4*a*c
  r1 = (-b+(d)**(1/2))/2*a
  r2 = (-b-(d)**(1/2))/2*a
  print("The roots are",r1,"and",r2)

a = int(input("Enter a value : "))
b = int(input("Enter b value : "))
c = int(input("Enter c value : "))
roots(a,b,c)
```

```python
#Fibanocci using recursion
def fib(n):
  if n<=1:
    return 1
  else:
    return fib(n-1)+fib(n-2)

n = int(input("Enter n value : "))
for i in range(1,n+1):
  print(fib(i))
```

```python
#WPP for finding GCD of two numbers
def gcd(a,b):
  fact_a = []
  fact_b = []
  hcf = []
  for i in range(1,a+1):
    if(a%i==0):
      fact_a.append(i)
  for i in range(1,b+1):
    if(b%i==0):
      fact_b.append(i)
  for i in fact_a:
    for j in fact_b:
      if (i==j):
        hcf.append(i)
  hcf.sort()
  return hcf[-1]

a = int(input("Enter first number : "))
b = int(input("Enter second number :"))
x = gcd(a,b)
print("The greatest common divisor of",a,"and",b,"is",x)
```
